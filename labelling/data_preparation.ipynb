```python
import pandas as pd
```

```python
original_data = pd.read_csv("../data/phone_ratings_cleaned.csv")
```

```python
original_data.head()
```

### Tiêu chí đánh nhãn:

- Rating (4-5): tích cực
- Rating (1-3): tiêu cực hoặc trung tính


```python
original_data["label"] = original_data["rating"].apply(
    lambda x: "positive" if x >= 4 else "negative"
)
```

```python
original_data
```

```python
sure_labels_path = "../data/labeled_phone_ratings_sure.csv"
unsure_labels_path = "../data/labeled_phone_ratings_unsure.csv"
```

```python
original_data[(original_data["rating"] <= 2)].to_csv(sure_labels_path, index=False) 
```

```python
unsure_data = original_data[original_data["rating"] >= 3]
mid_index = len(unsure_data) // 2

part1 = unsure_data.iloc[:mid_index]
part2 = unsure_data.iloc[mid_index:]

part1.to_csv("../data/unsure-labels/phone_ratings_unsure_part1.csv", index=False)
part2.to_csv("../data/unsure-labels/phone_ratings_unsure_part2.csv", index=False)
```

### Check lại các comment có rating từ 3-4


```python
double_check_data = original_data[(original_data["rating"] == 3) | (original_data["rating"] == 4)]
original_data = original_data[(original_data["rating"] != 3) & (original_data["rating"] != 4)]
```
