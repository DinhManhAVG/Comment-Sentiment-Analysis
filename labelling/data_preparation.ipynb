```python
import pandas as pd
```

```python
original_data = pd.read_csv("../data/processed_data/phone_ratings_cleaned.csv")
```

### Tiêu chí đánh nhãn:

- Rating (4-5): tích cực
- Rating (1-3): tiêu cực hoặc trung tính


```python
original_data
```

### Các sample có rating <= 2 thì khả năng cao là tiêu cực

```python
negative_comment_df = original_data[(original_data["rating"] <= 2)]
```

```python
negative_comment_df
```

```python
negative_comment_df["positive"] = 0
negative_comment_df["negative"] = 1
```

```python
sure_labels_path = "../data/sure_labels/labeled_phone_ratings_sure.csv"

negative_comment_df.to_csv(
    sure_labels_path,
    index=False,
)
```

### Lưu các sample có rating >= 3 vào các file *unsure_part_1.txt* và *unsure_part_2.txt* để đánh giá lại sau

```python
unsure_data = original_data[original_data["rating"] >= 3]
mid_index = len(unsure_data) // 2

part1 = unsure_data.iloc[:mid_index]
part2 = unsure_data.iloc[mid_index:]

part1.to_csv("../data/unsure_labels/phone_ratings_unsure_part1.csv", index=False)
part2.to_csv("../data/unsure_labels/phone_ratings_unsure_part2.csv", index=False)
```
